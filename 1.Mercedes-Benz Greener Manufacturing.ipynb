```python
#Extract the target variable
y=train_df['y']
```

```python
test_df=pd.read_csv(r"C:\Users\hp\Downloads\3.Chapter_ML_datas\ML_Projects\1.Mercedes-Benz Greener Manufacturing\test.csv")
```

```python
test_df.shape
```

```python
#1.If for any column(s), the variance is equal to zero, then you need to remove those variable(s)

var1=train_df.var(axis=0) #Checking the variance of train dataset
var1
```

```python
#Obtain the index of all the columns that has zero variance
list1=[]
for i,j in var1.items():
    if j==0:
        list1.append(i)
list1
```

```python
#Drop all the columns that has zero variance
train_df=train_df.drop(columns=list1,axis=1)
train_df.shape
```

```python
# #Checking the variance of test dataset
var2=test_df.var(axis=0)
var2
```

```python
list2=[]
for i,j in var2.items():
    if j==0:
        list2.append(i)
list2
```

```python
test_df=test_df.drop(columns=list2,axis=1)
test_df.shape
```

```python
#2.Check for null and unique values for test and train sets.
train_df.isnull().sum()
```

```python
train_missing_val=0
for col in train_df.columns:
    for item in train_df[col].isnull():
        if item==True:
            train_missing_val+=1
train_missing_val
```

```python
#Check whether there is any missing values in test dataset
test_missing_val=0
for col in test_df.columns:
    for item in test_df[col].isnull():
        if item==True:
            test_missing_val+=1
test_missing_val
```

```python
#3.Apply label encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#checking for the unique columns in the dataset
for col in train_df.columns:
    print(col,train_df[col].nunique())
    
```

```python
#Since only the first 10 columns have more number of unique values, apply label encoder from X0 to X8
X= train_df[['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']].apply(le.fit_transform)
```

```python
X.head()
```

```python
X.shape,y.shape
```

```python
#applying the label encoder to the test dataset as well
for col in test_df.columns:
    print(col,test_df[col].nunique())
    
```

```python
#Apply Label Encoder on the test data
test_df = test_df[['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']].apply(le.fit_transform)
```

```python
test_df.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
xtrain.shape,ytrain.shape,xtest.shape,ytest.shape
```

```python
#4.Perform dimensionality reduction.
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
```

```python
pca.fit(xtrain)
```

```python
xtrain_pca=pca.transform(xtrain)
```

```python
xtrain_pca.shape
```

```python
xtest_pca=pca.transform(xtest)
```

```python
xtest_pca.shape
```

```python
#5.Predict your test_df values using XGBoost.
from xgboost import XGBRegressor
xgb=XGBRegressor()
```

```python
xgb.fit(xtrain_pca,ytrain)
y_pred=xgb.predict(xtest_pca)
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print("Train acc:",xgb.score(xtrain_pca,ytrain))
print("Test acc:",xgb.score(xtest_pca,ytest))
```

```python
r2_score(ytest,y_pred)
```

```python
np.sqrt(mean_squared_error(ytest,y_pred))
```

```python
#predicting the test dataset values
testdf_pca=pca.transform(test_df)
```

```python
testdf_pca.shape
```

```python
target_test=xgb.predict(testdf_pca)
target_test
```

```python
print("Test acc:",xgb.score(testdf_pca,target_test))
```
